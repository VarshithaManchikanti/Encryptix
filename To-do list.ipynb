```python
def display_menu():
    print("\nTo-Do List Application")
    print("1. View To-Do List")
    print("2. Add To-Do Item")
    print("3. Remove To-Do Item")
    print("4. Exit")

def view_todo_list(todo_list):
    if not todo_list:
        print("\nYour to-do list is empty.")
    else:
        print("\nYour To-Do List:")
        for idx, item in enumerate(todo_list, start=1):
            print(f"{idx}. {item}")

def add_todo_item(todo_list):
    item = input("\nEnter a new to-do item: ")
    todo_list.append(item)
    print(f'"{item}" has been added to your to-do list.')

def remove_todo_item(todo_list):
    view_todo_list(todo_list)
    if todo_list:
        try:
            item_num = int(input("\nEnter the number of the item you want to remove: "))
            if 1 <= item_num <= len(todo_list):
                removed_item = todo_list.pop(item_num - 1)
                print(f'"{removed_item}" has been removed from your to-do list.')
            else:
                print("Invalid item number.")
        except ValueError:
            print("Please enter a valid number.")

def main():
    todo_list = []
    while True:
        display_menu()
        choice = input("\nChoose an option (1-4): ")
        if choice == '1':
            view_todo_list(todo_list)
        elif choice == '2':
            add_todo_item(todo_list)
        elif choice == '3':
            remove_todo_item(todo_list)
        elif choice == '4':
            print("Exiting the application. Goodbye!")
            break
        else:
            print("Invalid choice. Please choose a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
